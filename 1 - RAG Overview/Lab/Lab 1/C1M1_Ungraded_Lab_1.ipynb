## 1 - 列表

<a id='1-1'></a>
### 1.1 简介

列表（List）是 Python 中用于存储项目集合的内置数据类型。列表是有序的、可修改的，并且允许重复值。

````python
# 列表示例
l1 = ['RAG', 'is', 'awesome']
print(f"原始列表: {l1}")

# 添加一个元素
l1.append('!')
print(f"添加 '!' 后的列表: {l1}")

# 删除一个元素
l1.remove('awesome')
print(f"删除 'awesome' 后的列表: {l1}")
````

````python
# .remove 和 .append 会原地修改列表，且没有返回值
result = l1.append("this is a test")
print(result)
````

<a id='1-2'></a>
### 1.2 列表推导式

列表推导式提供了一种简洁的列表创建方式，使代码更具可读性与表达力。它可作为 `for` 循环的更简单替代方案。虽然使用列表推导式通常会带来[少量性能提升](https://stackoverflow.com/a/22108640)，但其主要优势仍在于可读性和简洁性，而不是显著的性能提升。

````python
# 使用列表推导式创建平方数列示例
squares = [x**2 for x in range(10)]
print(f"0 到 9 的平方数: {squares}（使用列表推导式）")


# 使用 for 循环实现同样的示例
squares_for_loop = []
for x in range(10):
    squares_for_loop.append(x**2)
print(f"0 到 9 的平方数: {squares_for_loop}（不使用列表推导式）")
````

````python
# 带条件的列表推导式
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(f"0 到 9 的偶数平方: {even_squares}（使用列表推导式）")

# 不使用列表推导式
even_squares_for_loop = []
for x in range(10):
    if x % 2 == 0:
        even_squares_for_loop.append(x**2)
print(f"0 到 9 的偶数平方: {even_squares_for_loop}（不使用列表推导式）")
````

## 2 - 字典

<a id='2-1'></a>
### 2.1 简介
字典（Dictionary）用于以 `key:value` 键值对形式存储数据，具有可修改性，且键不允许重复。

````python
# 字典示例
person = {
    'name': 'Alice',
    'age': 25,
    'city': 'New York'
}
print(f"person 字典: {person}")

# 访问某个值
print(f"Name: {person['name']}")

# 添加新的键值对
person['email'] = 'alice@example.com'
print(f"更新后的 person 字典: {person}")
````

<a id='2-2'></a>
### 2.2 `.items` 方法


你也可以使用 `.items` 方法遍历字典中的键值对：

````python
for key, value in person.items():
    print(f"Key: {key}\tValue: {value}")
````

如果你直接遍历字典本身，遍历到的是它的 `keys`：

````python
for val in person:
    print(val)
````

## 3 - f-strings

<a id='3-1'></a>
### 3.1 简介

Python 的 f-strings（格式化字符串字面量）可以在字符串中通过花括号 `{}` 嵌入表达式。你只需在字符串前加上 `f`，即可将其变为 f-string。


````python
# 基础 f-string 示例
name = "John"
age = 30
greeting = f"Hello, {name}. You are {age} years old."
print(greeting)
````

<a id='3-2'></a>
### 3.2 f-strings 实战

一种常见的做法是基于一组数据生成字符串。假设你有一个包含人员信息的字典列表，并希望生成如下格式的字符串：

```
Name: Alice Johnson, Age: 28, E-mail: alice.johnson@example.com, Location: New York, NY
Name: Michael Smith, Age: 34, E-mail: michael.smith@example.com, Location: Los Angeles, CA
...（持续输出）
```

````python
# 字典列表示例：这种结构在本课程中会经常出现！
people = [
    {
        "name": "Alice Johnson",
        "age": 28,
        "email": "alice.johnson@example.com",
        "location": "New York, NY"
    },
    {
        "name": "Michael Smith",
        "age": 34,
        "email": "michael.smith@example.com",
        "location": "Los Angeles, CA"
    },
    {
        "name": "Emily Davis",
        "age": 22,
        "email": "emily.davis@example.com",
        "location": "Austin, TX"
    },
    {
        "name": "John Brown",
        "age": 45,
        "email": "john.brown@example.com",
        "location": "Chicago, IL"
    },
    {
        "name": "Sarah Wilson",
        "age": 31,
        "email": "sarah.wilson@example.com",
        "location": "Seattle, WA"
    }
]
````

````python
# 首先，创建一个空列表用于存储句子：
t = []
# 遍历字典列表
for person_info_dict in people:
    # person_info_dict 是字典，你可以通过键访问其值
    layout_string = f"Name: {person_info_dict['name']}, Age: {person_info_dict['age']}, E-mail: {person_info_dict['email']}, Location: {person_info_dict['location']}"
    # 将目标信息字符串追加到列表中
    t.append(layout_string) # 在末尾添加一行
# 使用 .join 方法生成最终结果字符串
# .join 会按指定分隔符连接列表 't' 中的每个字符串，这里使用换行符 '\n'
formatted_string = "\n".join(t) 
print(formatted_string)
````

**重要**：访问字典中的值时，请记住键是字符串，因此需要一致地使用引号。在这个示例中，字典键使用单引号（`'`），而 f-string 外层字符串使用双引号（`"`）。如果引号使用不当，解析器将无法正确处理，进而报错。

````python
# 另一种根据参数创建字符串的方式如下：
template = "Name: {name}, Age: {age}, E-mail: {email}, Location: {location}"
t = []
for person_info_dict in people:
    # person_info_dict 是字典，你可以通过键访问其值
    # 使用 .format 方法生成并追加字符串
    layout_string = template.format(name = person_info_dict['name'], 
                         age = person_info_dict['age'], 
                         email = person_info_dict['email'], 
                         location = person_info_dict['location'])
    t.append(layout_string)
formatted_string = "\n".join(t)
print(formatted_string)
````
